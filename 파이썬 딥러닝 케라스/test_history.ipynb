```python
from keras.utils import np_utils
from keras.datasets import mnist
from keras.models import Sequential as sq
from keras.layers import Dense, Activation
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)

# 훈련 : _train, Y_train, 시험 셋 _test
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

maxn =50000

# 훈련_val , 검증_train
X_val = X_train[maxn:]
Y_val = Y_train[maxn:]
X_train = X_train[:maxn]
Y_train = Y_train[:maxn]

# [maxn][784] 형태로 변형
X_train = X_train.reshape(maxn, 784).astype('float32')/255.0
X_val = X_val.reshape(10000, 784).astype('float32')/255.0
X_test = X_test.reshape(10000, 784).astype('float32')/255.0

#(a,b) a의 범위중 랜덤으로 b개 선택
train_rand_idxs = np.random.choice(maxn, 700)
val_rand_idxs = np.random.choice(10000, 300)

# b 개 만큼 random 으로 섞인 데이터 생성
X_train = X_train[train_rand_idxs]
Y_train = Y_train[train_rand_idxs]
X_val = X_val[val_rand_idxs]
Y_val = Y_val[val_rand_idxs]

# 라벨링
'''
to_categorical
keras.utils.to_categorical(y, num_classes=None)
Converts a class vector (integers) to binary class matrix.
E.g. for use with categorical_crossentropy.
Arguments
y: class vector to be converted into a matrix (integers from 0 to num_classes).
num_classes: total number of classes.
Returns
A binary matrix representation of the input. The classes axis is placed last.
'''
Y_train = np_utils.to_categorical(Y_train)
Y_val = np_utils.to_categorical(Y_val)
Y_test = np_utils.to_categorical(Y_test)

# 모델 구성
model = sq()
model.add(Dense(units = 2, input_dim = 28*28, activation='relu'))
model.add(Dense(units = 10, activation='softmax'))

# 모델 엮기
model.compile(loss='categorical_crossentropy', optimizer ='sgd', metrics=['accuracy'])

hist = model.fit(X_train, Y_train,
                 epochs=1000,
                 batch_size=10,
                 validation_data=(X_val, Y_val))

```

```python

fig, loss_ax = plt.subplots()

acc_ax = loss_ax.twinx()

loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')

acc_ax.plot(hist.history['acc'], 'b', label='train acc')
acc_ax.plot(hist.history['val_acc'], 'g', label='val acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')

plt.show()




```
